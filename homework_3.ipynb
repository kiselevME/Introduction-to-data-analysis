```python
import numpy as np
import matplotlib.pyplot as plt

url = 'http://scipy-lectures.org/_downloads/populations.txt'
data = np.loadtxt(url, delimiter='	', dtype=int, skiprows=1)

years = data[:,0]
hare = data[:,1]
lynx = data[:,2]
carrot = data[:,3]
```

```python
#1 problem: Plot the populations for the years in the period. Add legend and axis labels to the plot.

plt.figure(figsize=(10,7))

plt.plot(years, hare, label='hare', c = 'blue')
plt.plot(years, lynx, label='lynx', c = 'red')
plt.plot(years, carrot, label='carrot', c = 'orange')
plt.xlabel('year')
plt.ylabel('quantity')
plt.legend()

plt.show()
```

```python
#2 problem: Assuming total population of hares and lynxes is 100%,
#plot stacked bar graph showing ratio of each specis for the years in the period.

total = hare + lynx
hare_ = hare/total[:]*100
lynx_ = lynx/total[:]*100

width = 0.8
plt.bar(years, label='hare', width = width, height = hare_)
plt.bar(years, label='lynx', width = width, height = lynx_, bottom = hare_)

plt.legend()
plt.xlabel('year')
plt.ylabel('percent')

plt.show()
```

```python
#Problem 3: Find the mean and std of the populations of each species. Plot the historgam of population for each
#species and show mean values with vertical line. Arrange 3 subplots in a row.

std_hare = np.std(hare)
std_lynx = np.std(lynx)
std_carrot = np.std(carrot)
mean_hare = np.mean(hare)
mean_lynx = np.mean(lynx)
mean_carrot = np.mean(carrot)

plt.figure(figsize=(16,5))

plt.subplot(1, 3, 1)
plt.title("Hare")
plt.xlabel("Years")
plt.ylabel("Quantity")
plt.barh(years, hare)
plt.axvline(mean_hare, c = 'red')
print('std_hare =', round(std_hare, 2))

plt.subplot(1, 3, 2)
plt.title("Lynx")
plt.xlabel("Years")
plt.ylabel("Quantity")
plt.barh(years, lynx)
plt.axvline(mean_lynx, c = 'red')
print('std_lynx =', round(std_lynx, 2))

plt.subplot(1, 3, 3)
plt.title("Carrot")
plt.xlabel("Years")
plt.ylabel("Quantity")
plt.barh(years, carrot)
plt.axvline(mean_carrot, c = 'red')
print('std_carrot =', round(std_carrot, 2))
```

```python
#Problem4: Find which year each species had the largest population. Mark these years in the plot of populations.

plt.figure(figsize=(10,7))

max_hare = np.max(hare)
max_lynx = np.max(lynx)
max_carrot = np.max(carrot)

max_year_hare = years[hare == max_hare][0]
max_year_lynx = years[lynx == max_lynx][0]
max_year_carrot = years[carrot == max_carrot][0]

plt.plot(years, hare, label='hare', c = 'blue')
plt.plot(years, lynx, label='lynx', c = 'red')
plt.plot(years, carrot, label='carrot', c = 'orange')
plt.xlabel('year')
plt.ylabel('quantity')
plt.legend()

plt.scatter(max_year_hare, max_hare, c = 'black', s = 70)
plt.scatter(max_year_lynx, max_lynx, c = 'black', s = 70)
plt.scatter(max_year_carrot, max_carrot, c = 'black', s = 70)

plt.show()
```

```python
#Problem 5: Find which species (hares or lynxes) has the largest population for each year. 
#Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

largest_population = np.where(hare > lynx, 'H', 'L')
print(largest_population)
labels='H','L'
plt.pie([np.sum(hare > lynx), np.sum(hare < lynx)], labels = labels, autopct='%0.1f%%')
plt.show()
```

```python
#Problem 6: Find which years any of the populations is above 50000. 
#Show time periods where populations are above 50000 in the plot of populations.

plt.figure(figsize=(12,7))

hare_b = hare[hare > 50000]
lynx_b = lynx[lynx > 50000]
carrot_b = carrot[carrot > 50000]
years_hare_b = years[hare > 50000]
years_lynx_b = years[lynx > 50000]
years_carrot_b = years[carrot > 50000]

plt.plot(years, hare, label='hare', c = 'blue')
plt.plot(years, lynx, label='lynx', c = 'red')
plt.plot(years, carrot, label='carrot', c = 'orange')
plt.scatter(years_hare_b, hare_b, c = 'Black', s = 80)
plt.scatter(years_lynx_b, lynx_b, c = 'Black', s = 80)
plt.scatter(years_carrot_b, carrot_b, c = 'Black', s = 80)

plt.xlabel('year')
plt.ylabel('quantity')
plt.legend()

plt.show()
```

```python
#Problem 7: Find the top 2 years for each species when they had the lowest populations.

hare_m = np.argsort(data[:, 1])
lynx_m = np.argsort(data[:, 2])

print("1 Hares: ", data[hare_m[0], 0])
print("2 Hares: ", data[hare_m[1], 0])
print("1 Lynxs: ", data[lynx_m[0], 0])
print("2 Lynxs: ", data[lynx_m[1], 0])
```

```python
#Problem 8: Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

plt.figure(figsize=(12,7))

plt.plot(years, np.insert(np.diff(data[:, 1]), 0, 0), label = 'hare', c = 'blue')
plt.plot(years, np.insert(np.diff(data[:, 2]), 0, 0), label = 'lynx', c = 'red')
plt.legend()
plt.grid()

plt.show()
```

```python
#Problem 9: Show population of hares vs carrots and hares vs lynxes.

plt.figure(figsize=(13,5))
plt.suptitle('Hares vs Carrots')

plt.subplot(1, 2, 1)
plt.plot(years, hare, c = 'blue')
plt.xlabel('years')
plt.ylabel('quantity')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(years, carrot, c = 'green')
plt.xlabel('years')
plt.ylabel('quantity')
plt.grid()


plt.figure(figsize=(13,5))
plt.suptitle('Hares vs Lynxes')

plt.subplot(1, 2, 1)
plt.plot(years, hare, c = 'blue')
plt.xlabel('years')
plt.ylabel('quantity')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(years, lynx, c = 'red')
plt.xlabel('years')
plt.ylabel('quantity')
plt.grid()
```
